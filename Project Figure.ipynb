```python
#import all that will be necessary
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
from scipy import stats

#import files for nutrient levels and cyanobacteria concentrations
nut_url = 'https://pasteur.epa.gov/uploads/10.23719/1503175/Nutrient_trend_data.xlsx'
cyano_url = 'https://pasteur.epa.gov/uploads/10.23719/1503175/Cyanobacteria_data.xlsx'
#select the sheet of on Excel that has the data I chose
cyano_sheet = 'SampleID_info_sum_cyano_CD'
nitro_sheet = 'TN_inflow'
#read the file. use sheet name to specifically read the one I want
#had difficulty with the excel files and how to read them. Used Google and YouTube to flip between the sheets.
cyanos = pd.read_excel(cyano_url,sheet_name = cyano_sheet)
nitro = pd.read_excel(nut_url, sheet_name = nitro_sheet,na_values='na')
#change the name year to Year in order to allign it with the Year on the cyanobacteria sheet
nitro = pd.DataFrame(nitro).rename(columns={'year':"Year"})
```

```python
#enter data frame
#cyanos is the read file
#sampleID is the column name needed to pull from
#needed to add the 'BVR' because file is full of other lakes and we only want BHR
#the contains function was a large issue to overcome for me as simple '.contains' was not working
#datetime to pull the years out
df = cyanos[cyanos['SampleID'].str.contains('BVR')]      
df['Date'] = pd.to_datetime(df[['Year','Month', 'Day']])
#needed to average the sum of the data for BVR per year. .mean helps us with that
#called on year because it is what needs to be averaged
df_no_loc = df[['Year', 'Sum of density_cells/ml']].groupby(df['Year']).mean()
#print(df_no_loc) to view the data frame
```

```python
#Want to merge the two data frames to make it easier when we pull parts of it out
#need to reset the index of both
#drop is to remove the old index
#inplace confirms to alter the data frame
#thank you Cristian for the help was having issue with the overall meaning of these functions and how to apply them.
#was originally not sure how to prep the data sets to merge them
df_no_loc.reset_index(drop= True, inplace=True)
nitro.reset_index(drop= True, inplace=True)
#merge the dataframes on the Year column
#this is why we had to rename earlier
data_table=pd.merge(df_no_loc, nitro, on='Year')
#print(data_table) to see that they are merged nicely
```

```python
#lines arent connecting on plot so we alter the code to drop the Nan values of both columns
#originally was having issues with the data points not connecting to eachother. By dropping the nan values this allows the recorded points to connect. Jackson was lots of help with this!
BVR = data_table[['Year','BVR']].dropna()
#plt.plot(BVR['Year'], BVR['BVR']) plotted it to make sure the lines are connecting now
```

```python
fig,axs=plt.subplots(figsize=(10,6)) #make plot dimensions
x = BVR['Year']
x2 = data_table['Year'] #set x-axis
#set the y axes
y1 = BVR['BVR']
y2 = data_table['Sum of density_cells/ml']
#plot the nitrogen in olive using y1. Make line and marker bigger to aid in the presentation
axs.plot(x,y1,marker='*', label = 'Total Nitrogen (mg/L)', color='olive',linewidth = 3.5, markersize = 15)
#labels
axs.set_ylabel('Total Nitrogen (mg/L)',fontsize = 15)
axs.set_xlabel('Years',fontsize = 15)
#set other axes
twinax = axs.twinx()
#use y2 because we're plotting the other side of graph. Make line and marker bigger to aid in the presentation
twinax.plot(x2,y2, marker='.',color='dodgerblue', label = 'Sum of density cells/ml', linewidth = 3.5, markersize = 15)
#labels change size of text to make easier to read
twinax.set_ylabel('Cyano sum of density cells/ml',fontsize = 15)
twinax.set_yscale('log')
#add title and grid
plt.title('Density of Cyanobacteria vs Total Nitrogen Levels - Brookville Lake (1988 - 2018)', fontsize=15)
axs.grid(True, linestyle=':', linewidth=0.5)
#legend and size
fig.legend(fontsize=9)
#increase the font size of the y axis and x axis for better reading
#used stackoverflow for these commands
axs.tick_params(axis='y',labelsize=13)
axs.xaxis.set_tick_params(labelsize=13)
twinax.yaxis.set_tick_params(labelsize=13)
```
